```python
import pandas as pd
# df = pd.read_csv("crime.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")
```

```python
cond1 = df["구분"] == "발생건수"
cond2 = df["구분"] == "검거건수"
df1 = df[cond1].iloc[:, 2:] # 발생건수만 가져오기 df2 = df[cond2].iloc[:, 2:] # 검거건수만 가져오기
```

```python
# 3) 연봉 / 근속연수 계산 후 세 번째로 높은 사람의 근속연수 (A)
# “근속연수 1년당 연봉이 얼마나 되는가” 같은 비율값 생성 예) 연봉 = 6000, 근속연수 = 3 이면, "연봉_근속연수" 값은 2000
```

```python
df_year = df.nlargest(3, '강력범죄') # '연봉_근속연수' 열이 없어 '강력범죄' 열로 대체하여 예시를 보여줍니다.
#  '강력범죄' 열을 기준으로 값이 큰 순서대로 상위 3개 행을 뽑는다.
```

```python
A = df_year.iloc[-1]["강력범죄"] # '근속연수' 열이 없어 '강력범죄' 열로 대체하여 예시를 보여줍니다.
```

```python
# 4) 연봉 / 만족도 계산 후 두 번째로 높은 사람의 교육참가횟수 (B)
# df["연봉_만족도"] = df["연봉"] / df["만족도"] # '연봉' 및 '만족도' 컬럼이 존재하지 않아 주석 처리합니다.
df_like = df.nlargest(2,'강력범죄') # '연봉_만족도' 컬럼이 존재하지 않아 '강력범죄'로 대체합니다.
```

```python
B = df_like.iloc[-1]["강력범죄"]
# 5) 결과 출력
result = A + B
print(result)
```
